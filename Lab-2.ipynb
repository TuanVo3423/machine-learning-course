```python
import numpy as np

# Create two vectors
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

# Dot product
dot_product = np.dot(vector1, vector2)
print("Dot Product:", dot_product)

# Cosine similarity
cosine_similarity = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
print("Cosine Similarity:", cosine_similarity)
# radian_12 = np.ar

# Vector addition
vector_sum = vector1 + vector2
print("Vector Addition:", vector_sum)
```

```python
from sympy import symbols, Eq, solve

# Define variables
x, y = symbols('x y')

# Define equations
equation1 = Eq(y, x + 1)
equation2 = Eq(y, 2*x + 1)

# Solve equations
solution1 = solve((equation1, equation2), (x, y))
print("Solution 1:", solution1)
```

```python
# Create two matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[2, 3], [4, 5]])

# Covariance
covariance_matrix = np.cov(matrix1, matrix2)
print("Covariance Matrix:")
print(covariance_matrix)

# Correlation
correlation_matrix = np.corrcoef(matrix1, matrix2)
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define stock prices
AppleStock = [70, 72, 80, 75, 69, 85]
MSStock = [120, 125, 130, 128, 110, 150]
GOStock = [200, 150, 120, 170, 250, 280]

# Create a matrix where each row represents a stock and each column represents a day
stock_prices = np.array([AppleStock, MSStock, GOStock])

# Calculate the correlation matrix
correlation_matrix = np.corrcoef(stock_prices)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', xticklabels=['Apple', 'Microsoft', 'Google'], yticklabels=['Apple', 'Microsoft', 'Google'])
plt.title('Correlation Matrix of Stock Prices')
plt.xlabel('Stocks')
plt.ylabel('Stocks')
plt.show()
```
