```python
import numpy as np
from sklearn.decomposition import PCA

# Tạo dữ liệu mẫu với 100 mẫu và 5 đặc trưng
x = np.random.randint(10, 100, size=(100, 5))

# Áp dụng PCA với số thành phần chính là 5
pca = PCA(n_components=5)
x_pca = pca.fit_transform(x)

# Chọn 3 thành phần chính tốt nhất
x_pca_selected = x_pca[:, :3]

print("Dữ liệu sau khi áp dụng PCA và chọn 3 thành phần chính tốt nhất:")
print(x_pca_selected)
```

```python
# Eigen Values and Eigen Vectors
import matplotlib.pyplot as plt
import numpy as np

# 2 columns, 1000 rows
CenterPoint = np.array([0,0])
# covarience
CovPoints = np.array([[0.6,0.2],[0.2,0.2]])
n = 1000
Points = np.random.multivariate_normal(CenterPoint,CovPoints,n).T
display(Points)
print(Points.shape)

A = np.cov(Points)
display(A)

# gia tri rieng
e, v = np.linalg.eig(A)

print("Eigen values :")
print(e)
print("Eigen vectors: ")
print(v)

plt.figure(figsize=(18,8))
plt.scatter(Points[0,:],Points[1,:], color = "b", alpha=0.2)
for ie, iv in zip(e, v.T):
    print(iv)
    plt.plot([0,3 * ie * iv[0]], [0,3 * ie * iv[1]],"r--", lw = 3)
    plt.show()
```
